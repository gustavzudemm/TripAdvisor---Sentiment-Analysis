```python
import pandas as pd
```

```python
'''Data Aggregation of Singapur Airline Reviews'''

df_1 = pd.read_csv('tripadvisor_airline_sq_batch_100.csv', index_col=0)
df_2 = pd.read_csv('tripadvisor_airline_sq_batch_200.csv', index_col=0)
df_3 = pd.read_csv('tripadvisor_airline_sq_batch_400.csv', index_col=0)
df_4 = pd.read_csv('tripadvisor_airline_sq_batch_600.csv', index_col=0)
df_5 = pd.read_csv('tripadvisor_airline_sq_batch_1000.csv', index_col=0)

frames = [df_1, df_2, df_3, df_4, df_5]
df = pd.concat(frames)

df.info
```

```python
'''Rating histogram'''
df[df['rating'] == 3.0]
```

```python
'''Load merged dataset of Singapore Airline Reviews into one csv-file'''

df.to_csv('tripadvisor_singapore_airline_reviews.csv', encoding='UTF-8')
```
